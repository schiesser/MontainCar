# Mountain Car Miniproject Notebook

```python
from RandomAgent import RandomAgent
from DQNAgent import DQNAgent
import matplotlib.pyplot as plt
import gymnasium as gym
from collections import namedtuple, deque
import numpy as np
env = gym.make('MountainCar-v0')
```

## Random Agent

### Single random agent

```python
simple_random = RandomAgent(env)
simple_random.run()
```

```python
print(f"Number of steps during episode = {simple_random.nb_step}")
```

```python
plt.plot(simple_random.observed_states_speed)
plt.title('Speed at each step of episode')
plt.xlabel('Step')
plt.ylabel('Speed')
plt.show()
```

```python
plt.plot(simple_random.observed_states_position)
plt.title('Position at each step of episode')
plt.xlabel('Step')
plt.ylabel('Position X')
plt.show()
```

```python
plt.plot(simple_random.observed_reward)
plt.title('Reward obtained at each step')
plt.xlabel('Step')
plt.ylabel('environnement reward')
plt.show()
```

### Multiple random agent 

```python
nb_episode = 100
nb_step_per_episode = np.zeros((100))
```

```python
for i in range(nb_episode):
    simple_random = RandomAgent(env, seed = i)
    simple_random.run()
    nb_step_per_episode[i] = simple_random.nb_step
```

```python
plt.scatter(range(nb_episode),nb_step_per_episode)
plt.title('Number of steps for each episode (with each a different initial position)')
plt.xlabel('Episode')
plt.ylabel('Number of steps')
plt.show()
```
